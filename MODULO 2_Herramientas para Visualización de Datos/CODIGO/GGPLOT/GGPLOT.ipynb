```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from plotnine import *

import os
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
reviews.head(3)
```

```python
top_wines = reviews[reviews['variety'].isin(reviews['variety'].value_counts().head(5).index)]
```

```python
df = top_wines.head(1000).dropna()

(ggplot(df)
  + aes('points', 'price')
   + geom_point())
```

```python
df = top_wines.head(1000).dropna()

(

    ggplot(df)
       + aes('points','price')
       + geom_point()
       + stat_smooth() ##media de todos
)
```

```python
df = top_wines.head(1000).dropna()

(

    ggplot(df)
       + aes('points','price')
       + aes(color='points')
       + geom_point()
       + stat_smooth()
)
```

```python
(

    ggplot(df)
       + aes('points','price')
       + aes(color='points')
       + geom_point()
       + stat_smooth()
       + facet_wrap('variety')
)
```

```python
(ggplot(top_wines)
    + aes('points')
    + geom_bar()
)
```

```python
pokemon = pd.read_csv("Pokemon.csv", index_col=0)\
                        .rename(columns=lambda x: x.replace(" ", "_"))
pokemon.head(3)
```

```python
(ggplot(pokemon)
  + aes('Attack', 'Defense')
   + geom_point())
```

```python
(
ggplot(pokemon)
       + aes('Attack','Defense')
       + aes(color='Legendary')
       + geom_point()
       + ggtitle('Pokemon Attack and Defense by Legendary Status')
)
```

```python
(ggplot(pokemon)
     + aes('Attack')
     + geom_histogram()
    + facet_wrap('~Generation')

)
```
