```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def minar(content,page,last):
    data=[]
    articles = content.find_all("div", class_="posting-card")
    
    articlesData = pd.DataFrame(columns=['l_tag', 'title', 'address','place', 'price1','price2', 'description', 'url','mainfeatures'])
    
    for node in articles:
        l_tag=node.find('span',class_= 'highlight-tag')
        title=node.find('a',class_='go-to-posting').text.strip('\n\t')
        address=node.find('span',class_= 'posting-location go-to-posting').text.strip('\n\t')
        address2=node.find('span',class_= 'posting-location go-to-posting')
        place=''
        for place in address2.find_all('span'):
          place = place.text.strip('\n\t')
        price1=node.find('span',class_= 'first-price')
        price2=node.find('span',class_= 'second-price')
        main_features=node.find('ul',class_='main-features')
        features=''
        for feature in main_features.find_all('li'):
          features = features+feature.find('b').text.strip('\n\t') + ','
        
        description=node.find('div',class_='posting-description').text.strip('\n\t')
        url=node.find('a',class_='go-to-posting').attrs['href']
        articleData=pd.DataFrame({
            'l_tag':l_tag.text if l_tag is not None else "",
            'title':title if title is not None else "",
            #'l_type':l_type.text if l_type is not None else "",
            'address':address if address is not None else "",
            'place': place if place is not None else "",
            'price1':price1.text if price1 is not None else "",
            'price2':price2.text if price2 is not None else "",
            'description':description if description is not None else "",
            'mainfeatures': features,
            'url':url if url is not None else ""
        },index=[last])
        last=last+1
        articlesData=articlesData.append(articleData)
    return articlesData,last
```

```python
df = pd.DataFrame(columns=['l_tag', 'title', 'address','place', 'price1', 'price2', 'description','url','mainfeatures'])
last=0
for pagina in range(1,200):
    url="https://urbania.pe/buscar/venta-de-casas?page="+str(pagina)
    #print(url)
    page = requests.get(url)
    
    soup = bs(page.content, 'html.parser')
    
    data,last=minar(soup,page,last)
    df=df.append(data)  
```
